```python
def load_substiles_dataset(path):
    episods = []
    dialogues = []
    files = glob(os.path.join(path, '*.ass'))
    
    if not files: 
        print(f"Aucun fichier trouvé dans le chemin : {path}")
    
    for file in files:
        print(f"Traitement du fichier : {file}")  
        episods.append(int(file.split('-')[1].split('.')[0]))
        
        with open(file, 'r') as f:
            lines = f.readlines()
            lines = lines[27:]
            rows = [",".join(line.split(',')[9:]) for line in lines]
        rows = [line.replace("\\N", ' ') for line in rows]
        script = " ".join(rows)       
        
        dialogues.append(script)
        df = pd.DataFrame.from_dict({'episode':episods,'script':dialogues})

    return df
```

```python
episode_1=df.iloc[0,1]
episode_1=sent_tokenize(episode_1)
labels=['dialogue','betrayal','battle','sacrifice','self development','hope','friend ship','love']
batch_size=20
for i in range (0,len(episode_1),batch_size):
    batch_sentences=" ".join(episode_1[i:i+batch_size])
    print(model_classifier(batch_sentences,labels, multi_label=True))
    
```

```python

#cette fonctionne travailsur une seul ligne du dataframe 
def get_theme_inference (df_p,batch_number=20):
    script_batch=[]
    theme_score_batch={}
    
    #tokenizer la ligne du dataframe 
    episode=sent_tokenize(df_p)
    
    #déviser la tliste des tokenz en batch
    for i in range (0   ,len(episode),batch_number):
        batch_sentences=" ".join(episode[i:i+batch_size])
        script_batch.append(batch_sentences)
        
    #appelle du modele pour classifier les theme 
    outputs=model_classifier(script_batch[:2],labels, multi_label=True)
    
    #pour chaque script batch on calcule le score de chaque theme  avec outputs est un list de  dict : 
    for output in outputs :
        for label,score in zip(output['labels'],output['scores']):
            if label not in theme_score_batch.values():
                theme_score_batch[label]=[]
            theme_score_batch[label].append(score)
            
    theme_score_batch = { key: np.mean(np.array(value)) for key ,value in theme_score_batch.items() }
        
    return theme_score_batch
```

```python
episode_1=df.iloc[0,1]
theme_score_batch=get_theme_inference(episode_1)
theme_score_batch=df.iloc[0:2,1]
print(theme_score_batch)
```

```python
print(theme_score_batch)
print("*******")
output_theme_df=theme_score_batch.apply(get_theme_inference)
print(output_theme_df)
print("********")
```
